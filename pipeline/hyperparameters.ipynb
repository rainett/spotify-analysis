```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
train_ds = pd.read_csv('/data/spotify_data_train.csv', index_col=0)
```

```python
y_column = 'Stream'
X_columns = ['Artist_encoded', 'Album_encoded', 'Danceability', 'Energy', 'Key',
             'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness',
             'Valence', 'Tempo', 'Duration_ms']
X_train = train_ds[X_columns]
y_train = train_ds[y_column]
```

```python
model = RandomForestRegressor()
param_grid = {
    'n_estimators': [800, 1000],
    'max_depth': [40],
    'max_features': [2, 4, 5],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2]
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, verbose=100)

grid_search.fit(X_train, y_train)

print("Найкращі гіперпараметри:", grid_search.best_params_)
print("Найкраща оцінка моделі:", grid_search.best_score_)
```

```python
r2_best_score = grid_search.best_score_
r2_best_params = grid_search.best_params_
print(f"score: {r2_best_score}, params: {r2_best_params}")
```
