```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("/data/Spotify_Youtube.csv", index_col=0)
```

```python
key_features = ['Danceability', 'Key', 'Energy', 'Loudness', 'Speechiness', 'Acousticness',
                'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms', 'Stream']
ds[key_features].describe()
```

```python
correlation_matrix = ds.corr()
missing_columns = set(ds.columns) - set(correlation_matrix.columns)
print("Відсутні колонки у матриці кореляцій:")
print(missing_columns)
# Візуалізація матриці кореляцій за допомогою теплової карти
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
